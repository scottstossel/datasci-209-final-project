```python
import pandas as pd
import altair as alt

currency_df = pd.read_csv('../data/raw/currency_data.csv')
currency_df.head(10)
```

```python
tourism_df = pd.read_csv('../data/preprocessed/formatted_tourists.csv')
tourism_df.head(10)
```

```python
# Convert tourism values to numeric
years = [str(year) for year in range(2009, 2020)]
for year in years:
    tourism_df[year] = pd.to_numeric(tourism_df[year], errors="coerce")

tourism_df["country"] = tourism_df["country"].str.strip().str.lower()

country_name_fixes = {
    "united states of america": "united states",
    "hong kong, china": "hong kong sar, china"
}

# Apply fixes to tourism dataset
tourism_df["country"] = tourism_df["country"].replace(country_name_fixes)

# Get top 10 countries by tourism in 2019
top_8_2019 = tourism_df.nlargest(7, "2019")["country"]

# Filter only these countries
tourism_top_8 = tourism_df[tourism_df["country"].isin(top_8_2019)]

# Reshape tourism data for Altair (Long Format)
tourism_melted = tourism_top_8.melt(id_vars=["country"], value_vars=years, 
                                     var_name="Year", value_name="Tourism")

color_scale = alt.Scale(scheme="category10")

# Create Tourism Line Chart
tourism_chart = alt.Chart(tourism_melted).mark_line().encode(
    x=alt.X("Year:O", title="Year", axis=alt.Axis(labelAngle=0)),
    y=alt.Y("Tourism:Q", title="Tourism Arrivals (thousands)"),
    color=alt.Color("country:N", title="Country", scale=color_scale),
).properties(
    width=800,
    height=700,
    title="Tourism Trends (2009-2019)"
).configure_title(
    fontSize=20
)

tourism_chart

# top_10_2019
```

```python
# Convert currency values to numeric
for year in years:
    currency_df[year] = pd.to_numeric(currency_df[year], errors="coerce")

currency_df["Country Name"] = currency_df["Country Name"].str.strip().str.lower()

# Filter only the top 10 countries
currency_top_8 = currency_df[currency_df["Country Name"].isin(top_8_2019)]

# Reshape currency data for Altair (Long Format)
currency_melted = currency_top_8.melt(id_vars=["Country Name"], value_vars=years, 
                                      var_name="Year", value_name="Exchange Rate")

# currency_df

# Create Exchange Rate Line Chart
exchange_chart = alt.Chart(currency_melted).mark_line().encode(
    x=alt.X("Year:O", title="Year", axis=alt.Axis(labelAngle=0)),
    y=alt.Y("Exchange Rate:Q", title="Exchange Rate (LCU per USD)"),
    color=alt.Color("Country Name:N", title="Country", scale=color_scale)
).properties(
    width=800,
    height=700,
    title="Exchange Rate Trends (2009-2019)",
).configure_title(
    fontSize=20
)

exchange_chart
```

```python
# Filter only these countries
tourism_top_8 = tourism_df[tourism_df["country"].isin(top_8_2019)].set_index("country")

# Filter only the top 8 countries in the currency dataset
currency_top_8 = currency_df[currency_df["Country Name"].isin(top_8_2019)].set_index("Country Name")

# Keep only 2009-2019 columns
tourism_top_8 = tourism_top_8[years]
currency_top_8 = currency_top_8[years]

# Compute correlation for each country separately
correlation_data = []
for country in top_8_2019:
    if country in tourism_top_8.index and country in currency_top_8.index:
        correlation_value = tourism_top_8.loc[country].corr(currency_top_8.loc[country])
        correlation_data.append({"country": country, "correlation": correlation_value})

# Convert to DataFrame
correlation_df = pd.DataFrame(correlation_data)
correlation_df["correlation"] = pd.to_numeric(correlation_df["correlation"], errors="coerce")
correlation_df["country"] = pd.Categorical(correlation_df["country"], categories=correlation_df["country"], ordered=True)

correlation_df["correlation"] = correlation_df["correlation"].fillna(0)

# Create a bar chart
bar_chart = alt.Chart(correlation_df).mark_bar().encode(
    x=alt.X("correlation:Q", title="Tourism & Exchange Rate Correlation"),
    y=alt.Y("country:N", title="Country"),  # Keep sorting but ensure country is categorical
    color=alt.Color("correlation:Q", title="Correlation", scale=alt.Scale(scheme='magma')),
    tooltip=["country", "correlation"]
).properties(
    width=600,
    height=400,
    title="Tourism & Exchange Rate Correlation by Country"
)


display(bar_chart)

# bar_chart

correlation_df.dtypes
```
