```python
import altair as alt
import numpy as np
import pandas as pd
```

```python
tourists_df = pd.read_csv('arrivals.csv', skiprows=1)
tourists_df = tourists_df.loc[:, ~tourists_df.columns.str.contains('Unnamed', na=False)]

tourists_df = tourists_df.drop(['C.','S.','C. & S.'], axis=1)
tourists_df = tourists_df.rename(columns={'Basic data and indicators': 'country'})
tourists_df.head(10)
```

```python
# fill empty cells with arrival data
tourists_df['country'] = tourists_df['country'].fillna(method='ffill')

# remove duplicate rows
tourists_df = tourists_df.dropna(subset=tourists_df.columns[1:], how='all').reset_index(drop=True)

tourists_df.head(10)
```
